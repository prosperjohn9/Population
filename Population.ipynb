```python
# Prompt for user's name
name = input("Please enter your name: ")

# Prompt the user for the end_year
end_year = int(input(name + ", Please enter a year for population estimation: "))

# Declare variables
population = 79510000
start_year = 2016

# Calculate how many seconds are in a year
sec_in_a_year = 365 * 24 * 60 * 60

#Calculate time elapsed in seconds
elapsed = (end_year - start_year) * sec_in_a_year

# Calculate current population
population = population + elapsed // 7 - elapsed // 13 + elapsed // 35

"""
You can also use assignment operation
population += elapsed // 7 - elapsed // 13 + elapsed // 35
You can also convert variables to string and concatenate
print('The estimated population will be ' + str(population) + ' in ' + str(end_year) + '.')

"""
# Print desired value
print("The estimated population will be ", population, " in ", end_year, ".", sep = "") # Separator specified as NULL (none)
```
